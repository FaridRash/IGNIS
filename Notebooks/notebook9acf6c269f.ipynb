```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
c1 = 3.7418e8
c2 = 1.4388e4
h = 6.626e-34
c = 3e8
```

```python
def compute_gsd(H, f, p):
  return (H * p / f) **2
```

```python
Tf = 433.15
```

```python
import cupy as cp

def compute_wighted_snr_gpu(Tg, Tf, theta_deg, R, tau_a):
    Af = 100  # 10*10
    Agsd = (R * pixel / focal_length) ** 2
    Ag = Agsd - Af

    theta_rad = cp.deg2rad(theta_deg)
    cos_theta = cp.cos(theta_rad)

    lambda_gpu = cp.array(lambda_range)  # must be pre-defined with cp

    M_lambda_g = c1 / (lambda_gpu**5 * (cp.exp(c2 / (lambda_gpu * Tg)) - 1))
    M_lambda_f = c1 / (lambda_gpu**5 * (cp.exp(c2 / (lambda_gpu * Tf)) - 1))

    E_lambda_g = epsilon * M_lambda_g
    E_lambda_f = epsilon * M_lambda_f

    coeff_g = (D**2 * t * cos_theta * Ag) / (4 * R**2 * h * c)
    coeff_f = (D**2 * t * cos_theta * Af) / (4 * R**2 * h * c)

    S_lambda_g = coeff_g * (lambda_gpu * E_lambda_g * tau_a * tau_0 * eta * delta_lambda)
    S_lambda_f = coeff_f * (lambda_gpu * E_lambda_f * tau_a * tau_0 * eta * delta_lambda)

    S_Ag = cp.sum(S_lambda_g)
    S_Af = cp.sum(S_lambda_f)

    S_prime_g = S_Ag / Ag
    S_prime_f = S_Af / Af
    S_prime = S_prime_f - S_prime_g

    N_prime = cp.sqrt(S_prime_f + S_prime_g)
    SNR_prime = S_prime / N_prime

    return cp.asnumpy(SNR_prime)
```

```python
Tg_range = cp.arange(283.15, 373.15, 1)
R_range = cp.arange(460000, 150000, -212.30)
tau_a = cp.arange(0.1, 0.5, .01)
theta_range = cp.arange(0, 90, 1)
Tg_range, R_range, tau_a, theta_range
```

```python
data = []
for i in Tg_range:
    for j in R_range:
        for k in tau_a:
            for l in theta_range:
                snr = compute_wighted_snr_gpu(i, Tf, l, j, k)
                data.append({
                    'Tg': i,
                    'Tf': Tf,
                    'theta': l,
                    'R': j,
                    'tau_a': k,
                    'weighted_SNR': snr
                })

# Convert to DataFrame and save
df = pd.DataFrame(data)
df.to_csv("snr_dataset_gpu.csv", index=False)
```
