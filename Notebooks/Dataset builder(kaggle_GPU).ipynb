```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import cupy as cp
```

```python
Tf_range = cp.arange(413.15, 434.15, 5)
Tg_range = cp.arange(283.15, 374.15, 10)
R_range = cp.arange(460000, 149999, -20665)
tau_a = cp.arange(0.1, 0.6, .05)
theta_range = np.arange(0, 26, 2)
len(Tg_range), len(R_range), len(tau_a), len(theta_range)
```
