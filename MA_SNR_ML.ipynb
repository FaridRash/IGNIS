```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/IGNIS/Dataset/MA_SNR.csv')
print(df.head())
```

```python
print(df['Altitude'].value_counts())
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Altitude'], kde=True)
plt.xlabel('Altitude (m)')
plt.ylabel('Frequency')
plt.title('Frequency Distribution of Altitude')
plt.tight_layout()
plt.show()
```

```python
df_filtered = df[df['Altitude'] > 370000]
print(df_filtered.head())
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_filtered, x='Altitude', y='SNR')
plt.xlabel('Altitude (m)')
plt.ylabel('SNR')
plt.title('SNR vs Altitude (Altitude > 370000 m)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='Month', y='SNR', data=df_filtered)
plt.title('SNR Distribution by Month')
plt.xlabel('Month')
plt.ylabel('SNR')
plt.grid(True)
plt.tight_layout()
plt.show()
```
